## Counting vocabulary

```python
print(bold, f"Length of text 3:", end)
print("\t", len(text3))

print(bold, "Words in alphabetical order:", end)
print("\t",sorted(set(text3))[:20])

print(bold, "Unique words:", end)
print("\t",len(set(text3)))

print(bold, "Distinct word percentage", end)
print("\t", len(set(text3)) / len(text3))

print(bold, "Occurences of word smote", end)
print("\t", text3.count("smote"))

print(bold, "Percentage of text that is the word 'a':", end)
print("\t", 100 * text4.count('a') / len(text4))

def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total

print(bold, "Lexical diversity of text3:", end)
print("\t",lexical_diversity(text3))

print(bold, "Lexical diversity of text5:", end)
print("\t",lexical_diversity(text5))

print(bold, "Percetnage of word 'a' in text4:", end)
print("\t",percentage(text4.count('a'), len(text4)))
```

### Exercise 17
Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import brown, stopwords
import string 


def most_freq(text, N = 50):
    fdist = nltk.FreqDist([w.lower() for w in text if w not in stopwords.words("english") and w[0] not in string.punctuation])
#     stopwords.words("english")
#     df = pd.DataFrame(fdist)
    return fdist.most_common(N)
most_freq(brown.words()[:1000])
```

### Exercise 18
Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk.corpus import brown, stopwords
import string 

def most_freq_bigrams(text, N = 50):
    SW = stopwords.words("english")
    fdist = nltk.FreqDist([(i.lower(),j.lower()) for i,j in list(nltk.bigrams(text)) 
                           if (i.lower() not in SW and j.lower() not in SW) and
                          i[0] not in string.punctuation and j[0] not in string.punctuation])
    return fdist.most_common(N)
most_freq_bigrams(brown.words()[:1000])
```
